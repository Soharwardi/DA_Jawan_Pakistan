```python
# Define the Employee class
class Employee:
    def __init__(self, name, number, address, salary, post, year_of_joining, education, experience):
        self.name = name
        self.number = number
        self.address = address
        self.salary = salary
        self.post = post
        self.year_of_joining = year_of_joining
        self.education = education
        self.experience = experience

    # Method to describe employee
    def describe_employee(self):
        print(f"Employee Name: {self.name}")
        print(f"Employee Number: {self.number}")
        print(f"Address: {self.address}")
        print(f"Salary: {self.salary}")
        print(f"Post: {self.post}")
        print(f"Year of Joining: {self.year_of_joining}")
        print(f"Education: {self.education}")
        print(f"Years of Experience: {self.experience}")
        print("")

    # Method to update year of experience
    def update_year_of_experience(self, new_experience):
        self.experience = new_experience

    # Method to increase salary
    def increase_salary(self, increment):
        self.salary += increment

# Create an instance
employee101 = Employee('John Doe', '0321-2040-5088', '123 Elm St', 60000, 'Engineer', 2015, 'Bachelors', 5)

# Print four attributes individually
print(employee101.name)
print(employee101.number)
print(employee101.address)
print(employee101.salary)

# Call methods
employee101.describe_employee()
employee101.increase_salary(5000)
employee101.update_year_of_experience(6)
employee101.describe_employee()

# Create 10 different instances
employees = [
    Employee('Alice', '0321-1234-5678', '456 Pine St', 70000, 'Manager', 2010, 'Masters', 10),
    Employee('Bob', '0321-2345-6789', '789 Oak St', 50000, 'Analyst', 2018, 'Bachelors', 3),
    Employee('Charlie', '0321-3456-7890', '101 Maple St', 80000, 'Director', 2005, 'PhD', 15),
    Employee('David', '0321-4567-8901', '202 Birch St', 55000, 'Consultant', 2017, 'MBA', 4),
    Employee('Eva', '0321-5678-9012', '303 Cedar St', 45000, 'HR', 2019, 'Bachelors', 2),
    Employee('Frank', '0321-6789-0123', '404 Ash St', 75000, 'Lead Engineer', 2011, 'Masters', 9),
    Employee('Grace', '0321-7890-1234', '505 Elm St', 60000, 'Scientist', 2014, 'PhD', 6),
    Employee('Hank', '0321-8901-2345', '606 Fir St', 62000, 'Designer', 2016, 'Bachelors', 5),
    Employee('Ivy', '0321-9012-3456', '707 Spruce St', 67000, 'Marketing', 2012, 'Masters', 8),
    Employee('Jack', '0321-0123-4567', '808 Willow St', 78000, 'Developer', 2008, 'Bachelors', 12)
]

# Call describe_employee() for each
for emp in employees:
    emp.describe_employee()

# Update some attributes
employees[0].increase_salary(3000)
employees[1].update_year_of_experience(4)
employees[2].increase_salary(4000)
employees[0].describe_employee()
employees[1].describe_employee()
employees[2].describe_employee()
```
